```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("TFC1LAIN.csv")
```

```python
data = data.set_index("Date").dropna()
```

```python
plt.plot(data['Futures'])
```

```python
rolling_mean = data['Futures'].rolling(window = 20).mean()
```

```python
plt.plot(data['Futures'], color = 'blue', label = 'Original')
plt.plot(rolling_mean, color = 'red', label = 'Rolling Mean')
plt.legend(loc = 'best')
plt.title('Rolling Mean of Futures')
plt.show()
```

```python
rolling_mean_full = data['Futures + Options'].rolling(window = 20).mean()
```

```python
plt.plot(data['Futures + Options'], color = 'blue', label = 'Original')
plt.plot(rolling_mean_full, color = 'red', label = 'Rolling Mean')
plt.legend(loc = 'best')
plt.title('Rolling Mean of Futures + Options')
plt.show()
```

```python
plt.plot(rolling_mean, color = 'red', label = 'Rolling Mean of Futures')
plt.plot(rolling_mean_full, color = 'blue', label = 'Rolling Mean of Futures + Options')
plt.legend(loc = 'best')
plt.title('Rolling Means')
plt.show()
```

```python
np.corrcoef(data['Futures'],data['Futures + Options'])
```

```python
#Demean 

np.mean(data['Futures'])
```

```python
plt.plot(data['Futures'] - np.mean(data['Futures']), color = "red", label = "Futures Demean")
plt.plot(data['Futures + Options'] - np.mean(data['Futures + Options']), color = "blue", label = "Futures + Options Demean")
plt.legend(loc = 'best')
plt.title('Asset Manager Net Standardized')
plt.show()
```
